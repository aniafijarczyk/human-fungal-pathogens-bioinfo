# DATA

```python
path = "/mnt/c/Users/aniaf/Projects/PC/GAPP/DATA/AFR_prediction_albicans/10_variants_5sets/"
```

```python
#dp = pd.read_csv(path + "bcftools_FLT2_snpEff_DP.tab", sep="\t", header=None, chunksize = 10 ** 5)
chunksize = 10 ** 5
filename = path + "bcftools_FLT2_snpEff_GT.tab"

R = []
i=0
for chunk in pd.read_csv(filename, sep="\t", header=None, chunksize=chunksize):
    print(i)
    gt_info = chunk.iloc[:, :5]
    gt_tab = chunk.iloc[:, 5:]
    
    df00 = gt_tab == '0/0'
    sum00 = df00.sum(axis=1)
    df01 = gt_tab.isin(['0/1','0/2','0/3','1/2','1/3','2/3'])
    sum01 = df01.sum(axis=1)
    df11 = gt_tab.isin(['1/1','2/2','3/3'])
    sum11 = df11.sum(axis=1)
    dfnn = gt_tab == './.'
    sumnn = dfnn.sum(axis=1)

    gt_info['REF_hom'] = sum00
    gt_info['ALT_hom'] = sum01
    gt_info['ALT_het'] = sum11
    gt_info['miss'] = sumnn

    gt_info['NGT'] = gt_info['REF_hom'] + gt_info['ALT_hom'] + gt_info['ALT_het']
    gt_info['ALT_hom_het'] = gt_info['ALT_hom'] + gt_info['ALT_het']
    gt_info['fALT_hom_het'] = gt_info['ALT_hom_het']/gt_info['NGT']
    
    gt_info = gt_info.rename(columns={0:'#CHROM', 1:'POS', 2:'REF', 3:'ALT',4:'TYPE'})
    gt_info['TYPE'] = gt_info['TYPE'].replace({'.':'SNP','1':'INDEL'})    

    R.append(gt_info)
    i+=1

dR = pd.concat(R, ignore_index=True)
print(dR.shape)
dR.head()
```

```python
dR.to_csv("11_08_get_GT.csv", sep=",", header=True, index=False)
```
